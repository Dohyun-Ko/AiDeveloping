```python
import numpy as np
```

```python
"""퍼셉트론 설계
      데이터 처리 용이함을 위해 행렬 사용

   신경망 설계
      각 퍼셉트론에서 다음 퍼셉트론으로 값을 전달하는 것을 시간 1로 침(모든 퍼셉트론이 동시에 작동하게 하기 위함)
      작동 방법
        1. 지정 숫자만큼 퍼셉트론 생성
        2. 퍼셉트론별로 

   퍼셉트론 클래스, 신경망 생성 클래스를 따로
"""
```

```python
def stepF(X):
  if np.sum(X) > 0.3:
    return 1
  else:
    return 0
```

```python
def ReLU(X):
  return np.maximum(0, np.sum(X))
```

```python
class Perceptron:
  def __init__(self, size, nSpN):
    self.wgt = np.random.rand(100)
    self.size = size
    self.nSpN = nSpN
    self.X = []
  def connectNeuron(self):
    self.cN = np.random.randint(0, self.size, self.nSpN)
  def input_(self, X):
    self.X = X
  def out(self):
    self.WX = self.X*self.wgt[:len(self.X)]
    return stepF(self.WX)
```

```python
class Network:
  def __init__(self, size, nSpN): #nSpN = the number of Synapse per Neuron
    self.size = size
    self.nSpN = nSpN
    self.neuron = []
  def newNeuron(self):
    n = Perceptron(self.size, self.nSpN)
    n.connectNeuron()
    self.neuron.append(n)
  def makeup(self):
    for i in range(self.size):
      self.newNeuron()
  def input_(self, X):
    self.neuron[0].input_(X)
  def tick(self):
    inputTable = []
    for i in range(self.size):
      inputTable.append('')
    for i in range(self.size):
      r = self.neuron[i].out()
      for j in range(self.nSpN):
        inputTable[self.neuron[i].cN[j]] += str(r)
        
    for i in range(self.size):
      temp = []
      if inputTable[i] != '':
        for j in inputTable[i]:
          temp.append(int(j))
        print(temp)
        self.neuron[i].input_(temp)
      
```

```python
#테스트
p = Perceptron(100, 5)
p.input(np.array([1,1,0,1]))
print(p.wgt)
print(p.WX)
print(p.out())
p.connectNeuron()
print(p.cN)
```

```python
#테스트
n = Network(20000, 10)
n.makeup()
n.neuron[0].cN
n.input_([1,1,2,0,-1])
n.tick()
```
